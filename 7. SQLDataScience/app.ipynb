```python
# !pip install sqlalchemy
import pandas 
from sqlalchemy import create_engine #With sqlalchemy, you can connect to various database engines (e.g., MySQL, PostgreSQL), create and execute SQL queries, and manage database objects like tables.

# Database connection parameters
db_host = 'localhost'
db_user = 'root'
db_password = ''
db_name = 'pima'
db_port = 3306  # default MySQL port

# Create a connection string
connection_string = f'mysql+pymysql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
# the string format should be: 'mysql+pymysql://username:password@host:port/database'

# Create an engine
engine = create_engine(connection_string)

# SQL query to fetch data
query = 'SELECT * FROM pima'

# Load data into a Pandas DataFrame
df = pandas.read_sql_query(query, engine)

# Display the DataFrame
df
```

```python
# now you can use the table to do analysis

# lets encode the outcome column: For example, if the 'Outcome' column originally contained values 'Yes' and 'No', they might be encoded as 1 and 0, respectively.
from sklearn.preprocessing import LabelEncoder # helps with categorical data encoding. LabelEncoder is a technique for converting textual labels (e.g., categories like "cat", "dog", "bird") into numerical values. 
encoder = LabelEncoder()
df['Outcome'] = encoder.fit_transform(df['Outcome'])
df
```
