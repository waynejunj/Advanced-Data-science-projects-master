```python
# install keras, tensorflow 

from keras.models import Sequential #It allows you to define a sequential neural network model, where layers are added one after another. 
from keras.layers import Dense #It represents a densely-connected layer, which is the most fundamental type of layer in neural networks. It performs matrix multiplication between the input data and a set of weights, followed by an activation function
```

```python
import pandas
# load the dataset
dataset = pandas.read_csv("pima.csv")
dataset
```

```python
# split into input (X) and output (y) variables
array = dataset.values
X = array[:,0:8]
y = array[:,8]
```

```python
# Split dataset into training set and test set
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)  # 70% training and 30% test
```

```python
# Neural network
model = Sequential()
# Input layers
model.add(Dense(12, input_dim=8, activation='relu'))
# hidden layers
model.add(Dense(8, activation='relu'))

# Output Layer
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

# loss='binary_crossentropy': This specifies the loss function used to measure the error between the model's predictions and the true labels (y_train).
# optimizer='adam': This defines the optimization algorithm used to update the weights of the network during training.
# metrics=['accuracy']: This specifies the metrics used to monitor the model's performance during training. 

# Fit your training data
model.fit(X_train, y_train, epochs=100, batch_size=64)
# epochs=100: This specifies the number of times the entire training dataset will be passed through the network. More epochs generally lead to better performance
# batch_size=64: This defines the batch size used during training. The model updates its weights after processing each batch. 
```

```python
# evaluate the model
score = model.evaluate(X_test, y_test,verbose=2)
#  Here, 2 indicates detailed output, including loss and other metrics accuracy.
print(score) #loss and accuracy
```
