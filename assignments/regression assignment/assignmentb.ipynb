```python
# Importing the necessary modules
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Getting the data
my_dataframe = pd.read_csv(r"C:\Users\Admin\Desktop\MODCOM2\datascience\datsciencepdfs\AdvancedLevel\machinelearning\csv\Customers.csv")
my_dataframe
```

```python
# Creating an array before inserting the data into my model
array = my_dataframe.values
array
X= array[0:500 , 2:6] #Avg. Session Length, Time on App,Time on Website, Length of Membership.
X
Y = array[0:500 , 6:8] #Yearly Amount spent
Y
```

```python
# import a model
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X , Y , test_size = 0.3 , random_state = 42)
```

```python
# We fit our trained data into our model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train , Y_train)

```

```python
predictions = model.predict(X_test)
print("PREDICTIONS = " , predictions)
```

```python
from sklearn.metrics import r2_score
print("Accuracy Score is: " , r2_score (predictions , Y_test) * 100)
```
