```python
data_frame['Gender'].replace({0:'Male',
                        1:'Female',
                        2:'Neutral'},
                         inplace = True)

data_frame['LiveOnCampus'].replace({0:'Off Campus',
                              1:'On Campus'},
                              inplace = True)

data_frame['HowCommute'].replace({1:'Walk',
                              2:'Bike',
                              3:'Car',
                              4:'Public Transit',
                              5:'Other'},
                              inplace = True)
data_frame
```

```python
data_frame['Gender'].fillna('Male' , inplace=True)
data_frame['Writing'].fillna(77.68 , inplace=True)
```

```python
# 7. Is there a difference in average sleep time between students who live on campus and those who commute?
from scipy.stats import stats
live_on_campus_sleep = data_frame[data_frame['LiveOnCampus'] == 'On Campus']['SleepTime'].mean()
commute_sleep = data_frame[data_frame['LiveOnCampus'] == 'Off Campus']['SleepTime'].mean()

tstat, pval = stats.ttest_ind(live_on_campus_sleep, commute_sleep)

print(f"Average Sleep Time (Live On Campus): {live_on_campus_sleep:.2f}")
print(f"Average Sleep Time (Commute): {commute_sleep:.2f}")
print(f"t-statistic: {tstat}, p-value: {pval}")

if pval < 0.05:
  print("There is a statistically significant difference in average sleep time between students who live on campus and commuters.")
else:
  print("There is not sufficient evidence to conclude a difference in sleep time based on living arrangements.")

```
