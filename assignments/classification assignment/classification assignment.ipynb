```python
import pandas as pd
```

```python
advert = pd.read_csv("Advertising.csv")
advert
```

```python
advert.isnull().sum()
```

```python
advert.describe()
```

```python
array = advert.values
array
X=array[0:200,1:4] #TV, Radio, Newspaper
X
Y=array[0:200,4:5]
Y
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,Y_train)
```

```python
prediction=model.predict(X_test)
print("Predictions:",prediction)

print("\n\n")

print("Expected outcome",Y_test)
```

```python
#checking the accuracy with R squared.
from sklearn.metrics import r2_score #it calculates the coefficient of the model
print("Accuracy=",r2_score(prediction,Y_test)*100)
```

```python
# 700,30,60
predict = [[700 , 30 ,60]]
answer=model.predict(predict)
print("The expected sales is:" ,answer )
```

```python
# If i drop radio advert will it affect my sales
from scipy.stats import stats
# answer = advert[['Tv' ,'Newspaper' , 'Radio ']].agg(['mean']) / brings an error
mean = 14.02
# The mean is from the describe()
#perform one-way ANOVA
t_statistic, p_value = stats.ttest_1samp(mean, popmean=0 )
if p_value > 0 :
    print("Doesn't affect the sales")
else:
    print("Affects the results")

```
