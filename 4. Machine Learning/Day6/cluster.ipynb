```python
#cleaning data: remove null/empty cells
customers.isnull().sum()
```

```python
#removing empty cells in column Annual Income
income_mean=customers["Annual Income"].mean()
income_mean
customers["Annual Income"].fillna(income_mean,inplace=True)
customers.isnull().sum()
```

```python
#remove empty cells in column gender, and replace with sting 'unknown'
customers["Gender"].fillna('unknown',inplace=True)
customers.isnull().sum()
```

```python
#replace empty cells in column spending score with its mean
spending_mean=customers["Spending Score"].mean()
spending_mean
customers["Spending Score"].fillna(spending_mean,inplace=True)
customers.isnull().sum()
```

```python
# step 5: Clustering (we are training data model not testing)
# we will be using K-Means algorithm when clusterings
#put your dataset in form of an array(unique list [])
array=customers.values
array
```

```python
X=array[0:200,2:5] #rows,columns
X
#we are splitting the dataset bcs K-means only works with numerical data types
#note we are using X only bcs we are performing an unsupervised learning algorithm
```

```python
#spliting into sets/clusters
from sklearn.cluster import KMeans as km
model=km(n_clusters=12,random_state=42)
model.fit(X)
#n_cluster groups our dataset(array) in groups of 12. the more the cluster/groups the better the selection
#random_state makes the dataset rows random everytime we cluster/group
```

```python
#next we find the mean/average of the groups (analysing) 
means = model.cluster_centers_
means
```

```python
#lets create a table for our groups
clusters=pandas.DataFrame(means,columns=['Age','Income','Spending Score'] ) #dataframe is used to create rows and columns
clusters
```

```python
#create a column group_number, for us to see the different rows with it group
customers["group_number"]=model.labels_
customers.head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x=customers["group_number"])
#y axis =number of customers
#title = distribution of group members
```
