```python
import pandas
data = pandas.read_csv("shared_articles.csv")
data.head(10)
```

```python
#do data cleaning
data.isnull().sum()
# drop column Unnamed: 13,14 and 15
# fill 'unknown' in null cell in column authorUserAgent,  authorRegion, and authorCountry
# replace en with english in column lang
```

```python
# We will only work with text colm, so lets put it ni form of an array(list)
X = data['text'].values
X
```

```python
# tokenization, remove stopwords, remove non-alphabetic words, 
#  we will utilize the sentence-transformers library to encode X into numerical embeddings. 
from sentence_transformers import SentenceTransformer #SentenceTransformer class allows you to load and use pre-trained sentence transformer models.
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens') #This combination allows the model to create numerical representations (embeddings) of sentences that capture not just the words themselves but also their meaning and relationships to each other. 

embeddings = model.encode(X, show_progress_bar = True) # to encode is change (a message) into code
# show_progress_bar displays a progress bar while the encoding process is happening (useful for larger datasets).
embeddings

# Hello How Are You => hello, How, Are, You => 5,1,3,6 => 18/4 => 3
```

```python
# in distilbert-base-nli-stsb-mean-tokens are several components that work together to create sentence embeddings
# DistilBERT-base: This is the underlying architecture of the pre-trained model used for sentence encoding
# Fine-tuned for NLI (Natural Language Inference):  helps the model capture semantic similarities and relationships between sentences. eg  relationship between two sentences, such as entailment (sentence B implies A), sentence A is the opposite of sentence B
# STS-B (Semantic Textual Similarity Benchmark): This dataset helps the model learn to encode sentences in a way that reflects their semantic meaning and how similar they are to each other.
# Mean Token Embeddings:Generates sentence embeddings by averaging individual word embeddings within a sentence.
```

```python
df = pandas.DataFrame(embeddings)
df
```

```python
# calculate the cosine similarity between all pairs of rows in a DataFrame (df) and then converts the resulting similarity matrix into a new DataFrame 
from sklearn.metrics.pairwise import cosine_similarity #calculates the cosine similarity between all pairs of rows in a given data structure.
cosine_sim_data = cosine_similarity(df) #calculates the cosine similarity between each row in df and every other row, resulting in a square matrix where each element represents the cosine similarity between the corresponding rows in the original DataFrame.

# pass above to be like a dataframe
cosine_data = pandas.DataFrame(cosine_sim_data)
cosine_data
```

```python
# now letrs retrieve similar news articles to a given news ID based on a pre-calculated cosine similarity matrix
def recommendations(newsid):
    index_recomendations = cosine_data.loc[newsid].sort_values(ascending=False).index.tolist()[1:10]
    # cosine_data.loc[newsid] retrieves the row in the cosine_data DataFrame that corresponds to the given newsid
    # loc[]- provides a convenient way to access specific rows, columns, or groups of rows/columns based on their labels (index or column names).
    # using sort_values() it sorts them in descending order
    # It then selects the indexes (which likely correspond to other news article IDs) from the sorted row, but excludes the first element ([1:10]) so we will get the first 9 similar articles
    # tolist converts dataframe or series or array to a python list
    print("Similar Indexes ", index_recomendations)
    
    news_titles = data['title'].loc[index_recomendations].values
    # uses the list of similar news article IDs (index_recomendations) to retrieve the corresponding titles from the title column in your original data DataFrame (data)
    #  The .values extracts the actual title strings from the resulting pandas Series object.

    print("Similar Articles", news_titles)


recommendations(3) #This line calls the recommendations function with the argument 3 and the function will find and print the titles of the 9 most similar articles to news article ID 3.
```
