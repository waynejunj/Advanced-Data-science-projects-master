```python
# Collaborative Filtering
import pandas
user_item_details = pandas.read_csv("file.csv")
user_item_details.head(20)
```

```python
movie_details = pandas.read_csv("Movie_Id_Titles.csv")
movie_details.head(20)
```

```python
# merge the two datasets
data = pandas.merge(user_item_details, movie_details, on = 'item_id')
data.head(20)
```

```python
# best rated movies. Find average rating per movie
data.groupby('title')['rating'].mean().sort_values(ascending=False).head(5)
```

```python
# worst rated movies.
# Find average rating per movie
data.groupby('title')['rating'].count().sort_values(ascending=False).tail(5)
```

```python
#get the number of ratings per movie
mean_ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
mean_ratings
mean_ratings['num_of_ratings'] = pandas.DataFrame(data.groupby('title')['rating'].count())
mean_ratings.head(5)
```

```python
# Find which movie did each user rate, and what was the rating  
# -PIVOT TABLE - an interactive way to quickly summarize large amounts of data
pivot   = data.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
pivot.head(5)
```

```python
# Lets correlate movies using the pivot table
# pivot - used to summarise, sort, reorganise, group, count, total or average data stored in a table
selected_movie = pivot['Godfather, The (1972)']

# find similar movies to this one based ratings
similar = pivot.corrwith(selected_movie) #calculates the Pearson correlation coefficient between the ratings of "The Godfather (1972)" and the ratings of all other movies in the pivot table
similar_df  = pandas.DataFrame(similar, columns = ['Correlations']) #creates a DataFrame named similar_df from the correlation results (similar). It adds a column named "Correlations" containing the correlation coefficients between "The Godfather" and other movies.
similar_df.sort_values('Correlations', ascending = False).head(50)
```

```python
# analyze movie similarity while considering the number of ratings each movie has received.
similar_df = similar_df.join(mean_ratings['num_of_ratings'])
similar_df.head(5)
```

```python
# Take only the one with over 200 ratings
similar_df[similar_df['num_of_ratings'] > 200].sort_values('Correlations', ascending = False).head(20)
```
