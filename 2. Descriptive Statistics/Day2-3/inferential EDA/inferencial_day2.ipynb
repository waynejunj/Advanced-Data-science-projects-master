```python
# This Code Converts the Exponent number to Decimal
from decimal import Decimal

def scientific_to_decimal(sci_str):
    try:
        # Convert the scientific notation string to a Decimal
        num = Decimal(sci_str)
        # Format the Decimal to a string without scientific notation
        decimal_str = format(num, 'f')
        return decimal_str
    except ValueError:
        return None

# Example usage
sci_str = "3.7955905916649034e-14"
decimal_value = scientific_to_decimal(sci_str)
print(decimal_value)
```

```python
# Does being an athlete affect a student's average study time?
athlete_study_time = data[data['Athlete'] == 'Yes']['StudyTime']
non_athlete_study_time = data[data['Athlete'] == 'No']['StudyTime']

# Perform an independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(athlete_study_time.dropna(), non_athlete_study_time.dropna(), nan_policy='omit')

# Interpret the results
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in average study time between athletes and non-athletes.")
else:
    print("There is no significant difference in average study time between athletes and non-athletes.")
```
