```python
import pandas
# Read data, manipulate your data, select columns, basic stats, check empties.
data = pandas.read_csv("cleanedDataset.csv")
data
```

```python
# To determine if the mean of a sample (data['Height']) is significantly different from a known population mean (popmean=198).
import scipy.stats as stats

#perform one sample t-test
stats.ttest_1samp(a=data['English'], popmean=60)
# T-Statistic: This value tells you how many standard deviations the sample mean is from the population mean. A high absolute value indicates a large difference.
# P-Value: This value helps you determine the significance of your results. If the p-value is less than 0.05, you reject the null hypothesis, indicating that the sample mean is significantly different from the population mean.
# Degrees of Freedom - is essential for determining the critical t-value from the t-distribution table and for interpreting the t-test results.

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(a=data['English'], popmean=50)

# Print the results
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value:.10f}")

if p_value < 0.05:
  print("We do not have sufficient evidence to say that the mean Engish score for all students is equal to inches 50 .")
else:
  print("sample mean is not significantly different from the population mean")

# We do not have sufficient evidence to say that the mean height for all students is equal to inches 198 .

```
